```python
import pandas as pd

df = pd.read_csv("Fake_Real.csv")
```

```python
df.head()
```

```python
df.drop(["Unnamed: 0", "subject"], axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['target'].value_counts()
```

```python
df['label_num'] = df['target'].map({
    "Fake":0,
    "True": 1
})
```

```python
df
```

```python
import spacy
! python -m spacy download en_core_web_lg
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
# Convert text to vector

df["vector"] = df["text"].apply(lambda text: nlp(text).vector)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(
    df.vector,
    df.label_num,
    test_size=0.33,
    random_state=42
)
```

```python
X_train.shape, X_test.shape
```

```python
import numpy as np

X_train_stack = np.stack(X_train)
X_test_stack = np.stack(X_test)
```

```python
X_train_stack.shape
```

```python
X_test_stack.shape
```

```python
from sklearn.naive_bayes import GaussianNB

classifier = GaussianNB ()
```

```python
classifier.fit(X_train_stack, y_train)
```

```python
y_pred = classifier.predict(X_test_stack)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```
