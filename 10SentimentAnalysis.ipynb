```python
import tensorflow as tf

from keras.datasets import imdb
```

```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(X_train, y_train), (X_test, y_test) = imdb.load_data()
```

```python
len(X_train[0])
```

```python
from keras.utils import pad_sequences

X_train = pad_sequences(X_train, maxlen=500, padding="post")
X_test = pad_sequences(X_test, maxlen=500, padding="post")
```

```python
print(len(X_train[0]))
print(len(X_train[1]))
print(len(X_train[3]))
```

```python
model = Sequential()
model.add(SimpleRNN(32, input_shape=(500,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(metrics=['accuracy'], optimizer = 'adam', loss='binary_crossentropy')
```

```python
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```
