```python
doc = nlp("dog cat banana aryan alsdgj")

for token in doc:
  print(token.text, "vector:", token.has_vector, "OOV:", token.is_oov)
```

```python
base_token = nlp("sandwich")
```

```python
base_token.vector.shape
```

```python
doc = nlp("bread sandwich burger tiger humen rice food ")

for token in doc:
  print(f"{token.text} <-> {base_token.text}:", token.similarity(base_token))
```

```python
def similar_words(base_word, word_to_compare):
  base_token = nlp(base_word)
  doc = nlp(word_to_compare)
  for token in doc:
    print(f"{token.text} <-> {base_token.text}:", token.similarity(base_token))
```

```python
similar_words("iphone", "apple samsung iphone dog cat banana")
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector
queen = nlp.vocab["queen"].vector

result = king - man + woman
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([result], [queen])
```
