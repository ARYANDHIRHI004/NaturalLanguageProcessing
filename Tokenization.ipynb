```python
import nltk as nlp
```

```python
# This will download required data for tokenization
nlp.download("punkt_tab")
```

```python
corpus = """I am a boy. I love to eat chicken. I like very spicy chicken curry with rice"""
```

```python
print(corpus)
```

```python
# Tokenization
# paragraph to sentance
from nltk.tokenize import sent_tokenize
```

```python
## docuent as sentance
document = sent_tokenize(corpus)
```

```python
print(document)
```

```python
for sentence in document:
  print(sentence)
```

```python
# Tokenization
# sentance to word
from nltk.tokenize import word_tokenize
```

```python
for sentence in document:
  print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
vocab = wordpunct_tokenize(corpus)
```
